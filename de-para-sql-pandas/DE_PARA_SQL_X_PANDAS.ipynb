### Bibliotecas

```python
import pandas as pd
import pandasql as ps
```

## Importando dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/gustavo-candidoo/data-science/main/insurance.csv")
```

### Filtrando

```python
df = df[['age', 'sex']]
```

## Select

A instru√ß√£o SELECT √© usada para selecionar dados de um banco de dados.

Os dados retornados s√£o armazenados em uma tabela de resultados, chamada conjunto de resultados.

```python
query = "SELECT age, sex FROM df LIMIT 6"
ps.sqldf(query,locals())
```

```python
df_select_pandas = df[['age', 'sex']]
df_select_pandas.head(6)
```

## Where

A cl√°usula WHERE √© usada para filtrar registros.

√â usado para extrair apenas os registros que atendem a uma condi√ß√£o especificada.

```python
query = "SELECT * FROM df WHERE age < 19 LIMIT 6"
ps.sqldf(query, locals())
```

```python
df_where_age_pandas = df[(df['age'] < 19)]
df_where_age_pandas.head(6)
```

```python
df_where_age_pandas = df[(df['age'] < 19)]
```

## Order by

A palavra-chave ORDER BY √© usada para classificar o conjunto de resultados em ordem crescente ou decrescente.

A palavra-chave ORDER BY classifica os registros em ordem crescente por padr√£o. Para classificar os registros em ordem decrescente, use a palavra-chave DESC.

```python
query = "SELECT * FROM df ORDER BY age DESC LIMIT 6"
ps.sqldf(query, locals())
```

```python
df_orderby_pandas = df.sort_values(by='age', ascending=False)
df_orderby_pandas.head(6)
```

## Group by

A instru√ß√£o GROUP BY agrupa linhas que possuem os mesmos valores em linhas de resumo, como "encontre o n√∫mero de clientes em cada pa√≠s".

A instru√ß√£o GROUP BY √© frequentemente usada com fun√ß√µes agregadas (COUNT(), MAX(), MIN(), SUM(), AVG()) para agrupar o conjunto de resultados por uma ou mais colunas.

```python
query = "SELECT AVG(age), sex FROM df GROUP BY sex"
ps.sqldf(query, locals())
```

```python
df_groupby_pandas = df.groupby('sex')['age'].mean()
```

```python
df_groupby_pandas = df.groupby('sex')['age'].mean().reset_index()
df_groupby_pandas
```

## Count

A fun√ß√£o COUNT() retorna o n√∫mero de linhas que corresponde a um crit√©rio especificado.

```python
query = "SELECT COUNT(age) FROM df GROUP BY age LIMIT 6"
ps.sqldf(query, locals())
```

```python
df_count_pandas = df['age'].value_counts().reset_index()
df_count_pandas.head(6)
```

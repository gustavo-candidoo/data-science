```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/gustavo-candidoo/data-science/main/datasets/gpa2.csv")
df.head()
```

```python
y = df['sat']
X = df.drop('sat', axis=1)
```

## Partitioning using K-Fold

```python
from sklearn.model_selection import KFold
```

```python
partitions = KFold(n_splits=5)
```

```python
for train_index, test_index in partitions.split(X.head(10)):
  print(train_index, len(train_index), test_index, len(test_index))
```

## Model training using K-Fold

```python
partitions = KFold(n_splits = 5, shuffle=True, random_state=22)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor

mse_train = []
mse_test = []

for train_index, test_index in partitions.split(X):

  X_train, X_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  model = DecisionTreeRegressor(max_depth = 2)
  model.fit(X_train, y_train)

  train_prob = model.predict(X_train)
  mse_train.append(mean_squared_error(y_train, train_prob))

  test_prob = model.predict(X_test)
  mse_test.append(mean_squared_error(y_test, test_prob))

result = pd.DataFrame({"MSE in training": mse_train, "MSE in test": mse_test})
result
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
df.sample(5)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Box Plot: Total de Gorjetas por Dia')
plt.xlabel('dia')
plt.ylabel('valor da conta')
plt.show()
```

```python
sns.countplot(x='day', data=df)
plt.title('Gráfico de Barras: Contagem de Gorjetas por Dia')
plt.xlabel('dia')
plt.ylabel('contagem')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Gráfico de Dispersão: Valor da Conta vs. Gorjeta')
plt.xlabel('valor da conta')
plt.ylabel('gorjeta')
plt.show()
```

```python
sns.lineplot(x='total_bill', y='tip', data=df)
plt.title('Gráfico de Linha: Gorjeta Sobre a Fatura Total')
plt.xlabel('valor da conta')
plt.ylabel('gorjeta')
plt.show()
```

```python
sns.kdeplot(df['total_bill'], shade=True)
plt.title('Gráfico de Densidade: Valor da Conta')
plt.xlabel('valor da conta')
plt.ylabel('densidade')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor - Correlação')
plt.show()
```
